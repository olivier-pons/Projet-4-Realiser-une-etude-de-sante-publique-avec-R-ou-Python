```python
# Il faut tout d'abord faire une jointure entre la table population et la table sous nutrition, en ciblant l'année 2017
sous_nutrition[['annee_debut','annee_fin']]=sous_nutrition['Année'].str.split('-',expand=True)
sous_nutrition['annee_debut'] = pd.to_numeric(sous_nutrition['annee_debut'], errors='coerce')
sous_nutrition['annee_fin'] = pd.to_numeric(sous_nutrition['annee_fin'], errors='coerce')
population['Année'] = pd.to_numeric(population['Année'], errors='coerce')
sous_nutrition['annee_mediane'] = (sous_nutrition['annee_debut'] + sous_nutrition['annee_fin']) / 2
sous_nutrition['mediane_sous_nutrition'] = round(sous_nutrition['sous_nutrition'] / 3)
pop_sous_nutrition = pd.merge(sous_nutrition,population,left_on=['annee_mediane','Zone'],right_on=['Année','Zone'])
pop_sous_nutrition = pop_sous_nutrition[pop_sous_nutrition['annee_mediane'] == pd.to_numeric('2017')]
display (pop_sous_nutrition)
```

```python
#Affichage du dataset
display(pop_sous_nutrition.head(100))
```

```python
#### Calcul et affichage du nombre de personnes en état de sous nutrition
somme_population_sous_nutrition = pop_sous_nutrition['mediane_sous_nutrition'].sum()
somme_population = pop_sous_nutrition['Population'].sum()
print('la population mondial est de : ' ,somme_population)
print ('nombre de personnes en état de sous-nutrition : ',somme_population_sous_nutrition , 'en 2017')
pourcentage_population_mondial = (somme_population_sous_nutrition / somme_population) * 100
print ('soit', round(pourcentage_population_mondial), '%')
```
